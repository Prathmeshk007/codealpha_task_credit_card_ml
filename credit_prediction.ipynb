```python
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

url = 'https://raw.githubusercontent.com/selinolgun/Credit-Score-Classification/master/train.csv'
df = pd.read_csv(url).head(200).dropna()

df = df.drop(columns=['ID','Customer_ID','Name','SSN','Month','Occupation','Type_of_Loan'], errors='ignore')

encoders = {}
for col in df.select_dtypes('object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
X = df.drop(columns=['Credit_Score'])
y = df['Credit_Score']
model = RandomForestClassifier(n_estimators=50).fit(X, y)
data = {'model': model, 'encoders': encoders, 'cols': list(X.columns)}
with open('model.pkl', 'wb') as f:
    pickle.dump(data, f)

```
